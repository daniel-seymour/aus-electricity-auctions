```python
import pandas as pd
import numpy as np
from nemosis import static_table, dynamic_data_compiler
import plotly.express as px
from datetime import datetime, timedelta

# Specify where we will be caching the raw AEMO data.
raw_data_cache = '/Volumes/T7/NEMO_data'
```

RRP (Regional Reference Price): Indicates that this price is region-specific and represents the market-clearing price for the lower regulation reserve.

LOWERREGRRP stands for Lower Regulation Reserve Regional Reference Price. It represents the price for lower regulation ancillary services in a specific region within the electricity market.

```python
from datetime import datetime, timedelta
import pandas as pd

# Define the start and end dates
start_date = datetime(2021, 11, 1)
end_date = datetime(2024, 10, 30)

# Initialize an empty DataFrame to store results with specified region names
columns = ['Date', 'NSW1_RRP', 'QLD1_RRP', 'SA1_RRP', 'TAS1_RRP', 'VIC1_RRP']
results_df = pd.DataFrame(columns=columns)

# Function to fetch and process data for a specific date
def process_date(date):
    # Define the 6:00 pm interval
    interval_start = date.replace(hour=18, minute=0, second=0)
    interval_end = interval_start + timedelta(minutes=5)
    
    # Fetch data for the specific interval
    price_data = dynamic_data_compiler(
        start_time=interval_start.strftime('%Y/%m/%d %H:%M:%S'),
        end_time=interval_end.strftime('%Y/%m/%d %H:%M:%S'),
        table_name='TRADINGPRICE',
        raw_data_location=raw_data_cache
    )
#     var = price_data[(price_data['REGIONID'] == 'NSW1')]
    
    rrps = {
    'NSW1_RRP': price_data[(price_data['REGIONID'] == 'NSW1')]['RRP'].values[0],
    'QLD1_RRP': price_data[(price_data['REGIONID'] == 'QLD1')]['RRP'].values[0],
    'SA1_RRP': price_data[(price_data['REGIONID'] == 'SA1')]['RRP'].values[0],
    'TAS1_RRP': price_data[(price_data['REGIONID'] == 'TAS1')]['RRP'].values[0],
    'VIC1_RRP': price_data[(price_data['REGIONID'] == 'VIC1')]['RRP'].values[0]
    }
    
    # Append results to the DataFrame with the correct column structure
    results_df.loc[len(results_df)] = [date.date()] + [rrps.get(region, None) for region in columns[1:]]

# Loop through each day in the date range
current_date = start_date
while current_date <= end_date:
    process_date(current_date)
    current_date += timedelta(days=1)

# Display the results
```

```python
temp_price_results = results_df.copy()
temp_price_results.head()
```

```python
import pandas as pd
import plotly.express as px
import plotly.io as pio



# Define column names
region_col = 'Region'
price_col = 'Price'

# List of region columns
region_columns = ['NSW1_RRP', 'QLD1_RRP', 'SA1_RRP', 'TAS1_RRP', 'VIC1_RRP']

# Melt the DataFrame to have all prices in a single column
df_melted = temp_price_results.melt(id_vars='Date', value_vars=region_columns, var_name=region_col, value_name=price_col)

# Plotting using Plotly Express
fig = px.histogram(df_melted, x=price_col, color=region_col, marginal="box", 
                   title="Price Distribution by Region", 
                   labels={price_col: "RRP (Price)", region_col: "Region"})

# Save the plot as an HTML file
pio.write_html(fig, '/Users/danielseymour/Documents/EC331-EME-thesis/prices_dist.html')  # Replace with your desired path and filename

# Show plot
fig.show()
```
